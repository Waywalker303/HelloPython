***1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0***

```python
x = df['yr_built'] - df['yr_renovated']  
df['delta_renovated'] = x.apply(lambda x: '0' if x <= 0 else x)
df.head(2)
```

```python
def year(x):
    data = x[:4]
    return data
```

```python
df['Yers'] = (df['date'].apply(year))
df.head(2)
```

```python
def month(x):
    data2 = x[4:6]
    return data2

df['month'] = (df['date'].apply(month))
df.head(2)
```
