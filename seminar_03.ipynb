***Задание 1***  
1.1 Создать новый признак Cpu_company, который будет содержать только название фирсы которая произвела CPU(intel,AMD и т.д)  
1.2 Создать новый признак Memory_Amount, который будет содержать только GB памятибез указания типа носителя (128,1024..)  
1.3 Создать новый признак Memory_Type, который будет содержать только тип ностителя (HDD,SDD ..)  
1.4 удалить признаки Memory & ScreenResoiution

```python
#если файл не загружаетсядобавить , encoding='cp1251'
df = pd.read_csv('data/laptop_price.csv', encoding='cp1251')
```

```python
df.head(2)
```

```python
#1.1
df['Cpu_company'] = df['Cpu'].apply(lambda x: x.split()[0])
df.head(2)
```

```python
#1.2
def func(x):
    memory = x.split()[0]
    if memory.endswith('GB'):
        resualt = memory.replace('GB','')
    elif memory.endswith('TB'):
        resualt = float(memory.replace('TB',''))*1024
    return int(resualt)
        
```

```python
## df['Memory_Amount'] = df['Memory'].apply(lambda x: x.split('GB')[0]) - в чисытых данных и простых условиях
df['Memory_Amount'] = df['Memory'].apply(func)
df.head(2)
```

```python
#1.3 Создать новый признак Memory_Type, который будет содержать только тип ностителя (HDD,SDD ..)  
def func_1(x):
    memory = x.split()[1:]
    return memory
```

```python
df['Memory_Amount'] = df['Memory'].apply(func_1)
df.head(2)
```

```python
# 1.4 удалить признаки Memory & ScreenResoiution
df.drop(columns = ['Memory','ScreenResolution'], inplace = True)
df.head(2)
```
